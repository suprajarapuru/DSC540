```python
import pandas as pd
import matplotlib.pyplot as plt
excel_file = 'world-population.xlsm'
file_input = pd.read_excel(excel_file)

plt.plot(file_input.Year, file_input.Population)

plt.xlabel('Year')
plt.ylabel('Population in billions')
plt.show
```
